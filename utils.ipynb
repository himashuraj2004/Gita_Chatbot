```python
import json
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
with open("bhagavad_gita_chapter_1.json","r", encoding="utf-8") as f:
    raw_data = json.load(f)["BhagavadGitaChapter"]

documents = []
for item in raw_data:
    text = f"Chapter {item['chapter']} Verse {item['verse']}:\n{item['text']}\n"
    documents.append(text)

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_text("\n".join(documents))
```

```python
hf_embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

Chroma.from_texts(chunks, embedding=hf_embed, persist_directory="gitadb")

print("Vector DB done")
```
